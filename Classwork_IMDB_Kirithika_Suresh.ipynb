Which movie made the highest profit? Who were its producer and director? Identify the actors in that film.

```python
df['profit']=df['revenue']-df['budget']
```

```python
df.head()
```

```python
max_profit=df['profit'].max()
```

The movie that made the highest Profit

```python
df_max=df[df['profit']==max_profit].reset_index(drop=True)
```

```python
df_max
```

Who were its producer and director?

```python
df_max['crew'][0]
```

```python
for i in eval(df_max['crew'][0]):
    if i['job']=='Producer' or i['job']=='Director':
        print(f'{i["job"]}:{i["name"]}')
```

Identify the actors in that film.

```python
df_max['cast'][0]
```

```python
for i in eval(df_max['cast'][0]):
    if i['gender']== 2 or i['gender']==1:
        print(i['name'])
```

2. This data has information about movies made in different languages. Which language has the highest average ROI (return on investment)?


```python
df['ROI']=(df['profit']/ df['budget']) * 100
```

```python
df.head()
```

```python
avg_roi_by_language = df.groupby('original_language')['ROI'].mean().sort_values(ascending=False)

# Find the language with the highest average ROI
highest_avg_roi_language = avg_roi_by_language.idxmax()
highest_avg_roi_value = avg_roi_by_language.max()
```

```python
highest_avg_roi_language
```

```python
highest_avg_roi_value
```
